### Sizes present and their Number

```python
data['size'].value_counts().plot(kind='barh')
plt.title('SIZE', fontdict = {'fontsize': 18})        
plt.xlabel('Amount')
plt.ylabel('sizes')
plt.show()
```

### Number of Male and Female waiters studied

```python
data['sex'].value_counts().plot(kind='bar',color='red')
plt.title('SEX', fontdict = {'fontsize': 18})        
plt.ylabel('Number')
plt.show()
```

### Relatioship Between Sex And Tip recieved

```python
data.groupby('sex')['tip'].sum().plot(kind = 'pie', figsize=(7,7))
plt.title('Relatioship Between Sex And Tip', fontdict = {'fontsize': 18}) 
plt.show()
```

#### since the number or male waiters are more than female, there's no real relationship between the 2 variables

### Days of the week on Record
